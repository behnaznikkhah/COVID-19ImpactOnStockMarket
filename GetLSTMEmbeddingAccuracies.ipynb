```
# GetLSTMEmbeddingAccuracies.ipynb

# Sai Madhuri Yerramsetti
# November 22, 2020
# Student Number: 0677671

# import required packages
import pandas as pd
import numpy as np
import os
import re
import sys
import matplotlib.pyplot as plt
from scipy.stats import randint
from sklearn.model_selection import train_test_split
from tensorflow.python.keras.preprocessing.text import Tokenizer
from keras.layers import Flatten, Dense, LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.optimizers import SGD
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import Adam
from keras.utils import np_utils
from keras import Sequential, optimizers
import tensorflow as tf

# To disable warnings which arise from chained assignment
pd.options.mode.chained_assignment = None

# Function to classify data using word embedding LSTM
def word_embedding_LSTM(X_train, X_test, y_train, y_test):
    
    to_append = []
    # fit the preprocessed data with tokenizer
    tokenizer = Tokenizer(num_words=30000)
    tokenizer.fit_on_texts(X_train)

    # get the vocabulary size       
    vocabulary_size = (len(tokenizer.word_index) + 1)
    print('Number of words: ', vocabulary_size)

    # convert text to sequences
    X_train_seqs = tokenizer.texts_to_sequences(X_train)
    X_test_seqs = tokenizer.texts_to_sequences(X_test)

    max_seq_length = 500
    print('Input length: ', max_seq_length)

    # pad the sequences
    X_train_pad = pad_sequences(X_train_seqs, maxlen=max_seq_length, padding='post')
    X_test_pad = pad_sequences(X_test_seqs, maxlen=max_seq_length, padding='post')

    # Convert class vector (integers) to binary class matrix
    Y_train = np_utils.to_categorical(y_train, 2)
    Y_test = np_utils.to_categorical(y_test, 2)

    # set the output embedding dimension as 100
    output_embedding_dimension = 100
    #Define optimizer to use in LSTM
    optimizer =tf.keras.optimizers.Adam(learning_rate=0.00001)

    # build the model, fit it on data and evaluate the results
    print('Build LSTM model with word embedding...')
    model = Sequential()
    model.add(Embedding(vocabulary_size, output_embedding_dimension, input_length=max_seq_length))
    model.add(LSTM(32, dropout = 0.2))
    #model.add(Dense(10, activation='relu')) 
    model.add(Dense(2))
    model.add(Activation('softmax'))

    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    print('.......Train the data using LSTM.......')
    history = model.fit(X_train_pad, Y_train, batch_size=32, epochs=60, verbose=0, validation_data=(X_test_pad, Y_test))

    # plot train and validation loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('train vs validation loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper right')
    plt.show()

    # Plot training & validation accuracy values
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('train vs validation accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['train', 'test'], loc='upper right')
    plt.show()

    loss, accuracy = model.evaluate(X_test_pad, Y_test, batch_size=32)
    print('Test loss:', loss)
    print('Test accuracy:', accuracy)

    to_append.extend([accuracy, loss])
    return(to_append)


if __name__ == "__main__":

  # initialize the variables
  file_path = 'final_data.csv'
  final_data = pd.read_csv(file_path)
  print("final data with headlines and stock market labels: \n", final_data.head(15))

  # Get countries
  country_list = final_data.Country.unique().tolist()

  # Create a new dataframe to save accuracies data
  column_names = ["LSTM_accuracy", "LSTM_loss"]
  acc_data = pd.DataFrame(columns = column_names)

  for country in country_list:
    print(".................Classification started for", country,"..................")
    # filter the data countrywise
    model_data = final_data[final_data['Country'] == country]
    print("Country-wise data: \n", model_data.head(5))
    print("Dimensions of country-wise data: ", model_data.shape)
    model_data = model_data.reset_index(drop=True)

    # Splitting the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(model_data['News'], model_data['Label'], test_size=0.2, random_state=42)        

    # vectorize data with word embedding and classify with LSTM
    to_append = word_embedding_LSTM(X_train, X_test, y_train, y_test)

    # add the accuracy values of LSTM as rows of acc_data dataframe created
    acc_data.loc[len(acc_data)] = to_append
    print(acc_data.head())

  # change countries list to series and add as new column in acc_data    
  countries = pd.Series(country_list)        
  acc_data['Country'] = countries.values        

  print(acc_data.head(14))
  #save data to csv
  acc_data.to_csv(r'acc_data_LSTMEmbedding.csv', index = False, header=True)
  from google.colab import files
  files.download("acc_data_LSTMEmbedding.csv")
  
```
